```python
import numpy as np
import multiprocessing
import time
import threading
import os

matrix_size = 500
num_matrices = 500

num_cores = multiprocessing.cpu_count()
max_threads = 2 * num_cores

print("Number of CPU cores:", num_cores)
print("Testing threads up to:", max_threads)

constant_matrix = np.random.rand(matrix_size, matrix_size)
matrices = [np.random.rand(matrix_size, matrix_size) for _ in range(num_matrices)]

def task(mat):
    np.dot(mat, constant_matrix)

thread_counts = list(range(1, max_threads + 1))
times = []

for numberOfThreads in thread_counts:

    startTime = time.time()
    activeThreads = threading.active_count()

    print(f"\nRunning with {numberOfThreads} threads...")

    for mat in matrices:

        t = threading.Thread(target=task, args=(mat,))
        t.start()

        while threading.active_count() - activeThreads > numberOfThreads:
            time.sleep(0.01)

    while threading.active_count() > activeThreads:
        print("Thread still running (left %d)..." %
              (threading.active_count() - activeThreads))
        time.sleep(0.05)

    endTime = time.time()
    total_time = (endTime - startTime) / 60
    times.append(total_time)

    print(f"Threads = {numberOfThreads}, Time = {total_time:.2f} min")

print("\n========== FINAL RESULTS ==========")
print("Threads   :", thread_counts)
print("Time(min) :", [round(t, 2) for t in times])
print("==================================")
```
